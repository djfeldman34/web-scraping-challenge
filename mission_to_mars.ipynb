```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from splinter import Browser
import html5lib
import time
```

```python
url1 = "https://mars.nasa.gov/news/"
```

```python
r = requests.get(url1)
print(r)
```

```python
soup = bs(r.content, 'html5lib') 
print(soup.prettify()) 
```

```python
x = soup.find('div', 'content_title')
title1 = x.text
print(title1)
```

```python
y = soup.find('div', 'rollover_description_inner')
content1 = y.text
print(content1)
```

```python
url2 = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
```

```python
r2 = requests.get(url1)
print(r2)
```

```python
soup2 = bs(r2.content, 'html5lib') 
print(soup.prettify()) 
```

```python
with Browser("chrome") as browser:
    # Visit URL
    browser.visit(url2)
    # Find and click the 'search' button
    button = browser.find_by_id('full_image')
    # Interact with elements
    button.click()
    time.sleep(5)
    button2 = browser.find_by_text('more info     ')
    button2.click()
    time.sleep(5)
    img1 = browser.find_by_id('page')
    img2 = img1.find_by_tag('figure')
    img3 = img2.find_by_tag('a')
    img3.click()
    time.sleep(5)
    imagepath = browser.url
```

```python
print(imagepath)
```
